```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.pipeline import Pipeline
import sklearn.metrics as metrics
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier,GradientBoostingClassifier, ExtraTreesClassifier, AdaBoostClassifier
from imblearn.over_sampling import SMOTE
from sklearn.metrics import confusion_matrix, plot_confusion_matrix,plot_roc_curve
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.optimizers import Adam, SGD
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.callbacks import EarlyStopping
from sklearn.preprocessing import StandardScaler
import seaborn as sns
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
```

```python

#Reading csv from feature engineering
train = pd.read_csv('./data/train_modelready.csv')
train
```

```python
train
```

### Baseline model - Preprocessing

```python
train['WnvPresent'].value_counts(normalize=True)
```

Wnv not Present is obviously oversample (94.7% of the data). However, we will still proceed with dumping data into our baseline model for benchmark.

```python
#Define X,y
X = train.drop(columns=['WnvPresent'])
y = train['WnvPresent']

#Train/test split data
X_train, X_test, y_train, y_test = train_test_split(X,
                                                          y,
                                                          stratify=y,
                                                          test_size=0.20,
                                                          random_state=42,
                                                          shuffle = True)
```

```python
print(X_train.shape, y_train.shape, X_test.shape, y_test.shape)
```

```python
#Create a Randomforest Classifier as baseline model
rf=RandomForestClassifier(n_estimators=100)
#fit train into model
rf.fit(X_train,y_train)
#predict
y_pred=rf.predict(X_test)

print('Train Score: ', rf.score(X_train, y_train), 'Test Score: ', rf.score(X_test, y_test))
print('ROC_AUC Score: ',metrics.roc_auc_score(y_test, y_pred.ravel()))
```

Baseline model acts as our null hypothesis, whatever result that fails under baseline model will be rejected.
Given that there is oversampling, we are still able to achive 57.07% ROC-AUC for train-test prediction and this will be our baseline model.

```python
#Plot for accuracy visualization.
plot_confusion_matrix(rf, X_test, y_test)
```

### Add-on with SMOTE to resolve oversampling

```python
smote = SMOTE()
#Resampling
X_smote, y_smote = smote.fit_resample(X_train, y_train)

print(X_smote.shape, y_smote.shape)
print('Positive Class with SMOTE', sum(y_smote)/len(y_smote) * 100, ' %')
```

Now Train data has equal samples 

### Putting smote back to Random forest

```python
#fit train into model
rf.fit(X_smote,y_smote)
#predict
y_pred=rf.predict(X_test)

print('Train Score: ', rf.score(X_train, y_train), 'Test Score: ', rf.score(X_test, y_test))
print('ROC_AUC Score: ',metrics.roc_auc_score(y_test, y_pred.ravel()))
```

Random forest with SMOTE has better ROC_AUC of 61.13% than baseline model. This shows that with SMOTE, it improves the prediction. Therefore, we proceed further to implement SMOTE into more classification models.

### Tensorflow - Keras

```python
ss = StandardScaler()
X_train_sc = ss.fit_transform(X_smote)
X_test_sc = ss.transform(X_test)
```

```python
print(X_train_sc.shape,X_test_sc.shape)
```

```python
model = Sequential()
n_input = X_train_sc.shape[1]
model.add(Dense(32, # How many nodes/neurons do you want in your first layer.
                input_shape=(n_input,),
                activation='relu'))
model.add(Dense(80))
model.add(Dropout(0.5))
model.add(Dense(30))
model.add(Dense(10))
model.add(Dense(5))
model.add(Dense(1, activation='sigmoid'))
model.summary()

# Compile
opt = Adam(learning_rate=0.001)
model.compile(loss='binary_crossentropy', optimizer=opt, metrics=['accuracy'])

# Model Fit
early_stop = EarlyStopping(monitor='val_loss', 
                           min_delta=0, 
                           patience=30, 
                           mode='auto')

history = model.fit(
    X_train_sc,
    y_smote,
    validation_data=(X_test_sc, y_test),
    epochs=50,
    batch_size=512,
    callbacks=[early_stop] # Add your callbacks here!
)
```

```python
history.history['accuracy'][-1], history.history['val_accuracy'][-1]
```

```python
plt.plot(history.history['accuracy'], label='Train accuracy')
plt.plot(history.history['val_accuracy'], label='Val accuracy')
plt.legend();
```

```python
train_loss = history.history['loss']
test_loss = history.history['val_loss']

plt.figure(figsize=(12, 8))
plt.plot(train_loss, label='Training loss')
plt.plot(test_loss, label='Testing loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend();
```

```python
pred = model.predict(X_test_sc).round()
cf_matrix = confusion_matrix(y_test, pred)
cf_matrix
```

```python
ax = sns.heatmap(cf_matrix, annot=True, 
             cmap='Blues', fmt='g')

ax.set_title('Seaborn Confusion Matrix with labels\n\n');
ax.set_xlabel('\nPredicted Values')
ax.set_ylabel('Actual Values ');

## Ticket labels - List must be in alphabetical order


## Display the visualization of the Confusion Matrix.
plt.show()
```

```python
print('Keras ROC_AUC score :', metrics.roc_auc_score(y_test, pred.ravel()))
```

Basic Keras learning has achieved 59.15% ROC_AUC. However, it is lower than Random forest with SMOTE. Thus, we will reject this model.

```python
from tensorflow.keras.regularizers import l2

# Model Architecture
model_l2 = Sequential()

n_input = X_train_sc.shape[1]

model_l2.add(Dense(60, input_dim=n_input, activation='relu', kernel_regularizer=l2(0.01)))
model_l2.add(Dense(80))
model_l2.add(Dropout(0.5))
model_l2.add(Dense(30))
model_l2.add(Dense(10, activation='relu', kernel_regularizer=l2(0.01)))
model_l2.add(Dense(1, activation='sigmoid', kernel_regularizer=l2(0.01)))

# Compile
opt = Adam(learning_rate=0.001)
model_l2.compile(loss='binary_crossentropy', optimizer=opt, metrics=['accuracy'])

# Fit
history_l2 = model_l2.fit(
    X_train_sc,
    y_smote,
    validation_data=(X_test_sc, y_test),
    epochs=50,
    batch_size=32,
    callbacks=[early_stop]
)
```

```python
train_loss_l2 = history_l2.history['loss']
test_loss_l2 = history_l2.history['val_loss']

plt.figure(figsize=(12, 8))
plt.plot(train_loss, label='Training loss')
plt.plot(test_loss, label='Testing loss')
plt.plot(train_loss_l2, label='L2 Training loss')
plt.plot(test_loss_l2, label='L2 Testing loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend();
```

```python
history_l2.history['accuracy'][-1], history_l2.history['val_accuracy'][-1]
```

```python
pred_l2 = model_l2.predict(X_test_sc).round()
cf_matrix_l1 = confusion_matrix(y_test, pred_l2)
ax = sns.heatmap(cf_matrix_l1, annot=True, 
             cmap='Blues', fmt='g')

ax.set_title('Seaborn Confusion Matrix with labels\n\n');
ax.set_xlabel('\nPredicted Values')
ax.set_ylabel('Actual Values ');

## Ticket labels - List must be in alphabetical order


## Display the visualization of the Confusion Matrix.
plt.show()
```

```python
print('Keras with l2 regularization ROC_AUC score :', metrics.roc_auc_score(y_test, pred_l2.ravel()))
```

Keras with L2 regularization achieved slight better than Random forest with 61.56% ROC_AUC. We will consider this model and proceed with more classifiers with Gridsearch and pipeline

#### Gridsearch with pipeline

Define models and parameters

```python
# Logistic Regression
logreg = LogisticRegression(
    solver='liblinear',
    max_iter=1000, 
    random_state=42,
)

logreg_params = {
    'clf__penalty': ['l1', 'l2'],
    'clf__C': [0.1, 0.5, 1, 1.5, 2.5]
}

# Gradient Boost
gb = GradientBoostingClassifier(random_state=42)

gb_params = {
    'clf__learning_rate': [0.05, 0.1, 0.15],
    'clf__max_depth': [2, 3,4]
}

# XGBoost
xgb = XGBClassifier(
    use_label_encoder=False,
    eval_metric='auc',
    objective='binary:logistic',
    random_state=42,
    n_jobs=-1
)

xgb_params = { 
    'clf__max_depth': [3, 4,5], #Depth of tree (higher value implies likely overfit)
    'clf__gamma' : [0.1, 0.2 , 0.3], #Min loss reduction required for further partition
    'clf__learning_rate' : [0.1, 0.2], #step size feature weight shrinkage
}

# Random Forest
rf = RandomForestClassifier(random_state=42, n_jobs=-1)

rf_params = {
    'clf__n_estimators': [100, 150, 200],
    'clf__max_depth': [4, 5, 6]
}

# Extra Trees
et = ExtraTreesClassifier(random_state=42, n_jobs=-1)

et_params = {
    'clf__n_estimators': [50, 75, 100],
    'clf__max_depth': [4, 5, 6]
}

#LightGBM
lgbm = LGBMClassifier()

lgbm_params = {
    'clf__learning_rate': [0.01,0.1,0.05],
    'clf__n_estimators': [16,24,32],
    'clf__num_leaves': [16,24,32], # large num_leaves helps improve accuracy but might lead to over-fitting
    'clf__boosting_type' : ['gbdt'], 
    'clf__objective' : ['binary'],
    'clf__max_bin':[255], # large max_bin helps improve accuracy but might slow down training progress
    'clf__random_state' : [500,1000,1500],
    'clf__colsample_bytree' : [0.65,0.7,0.8],
    'clf__subsample' : [0.7,0.8],
    'clf__reg_alpha' : [1.2,1.5,1.6],
    'clf__reg_lambda' : [1.4,1.5,1.6]
    }

ada = AdaBoostClassifier()

ada_params = {
    'clf__n_estimators':[50,100,150], 
    'clf__learning_rate':[1,0.1,0.01]
}
# List of classifiers & params
classifiers = [
    (lgbm, lgbm_params),
    (logreg, logreg_params),
    (gb, gb_params),
    (xgb, xgb_params),
    (rf, rf_params),
    (et, et_params),
    (ada, ada_params)
]
```

```python
result = []
```

#### Define functions for Gridsearch

```python
def model(clf, clf_params):
    
    # Instantiate pipeline
    pipe = Pipeline([
        ('ss', StandardScaler()),
        ('clf', clf)
    ])
    
    # Gridsearch for best estimator
    grid = GridSearchCV(
        pipe,
        param_grid=clf_params,
        scoring='roc_auc',
        verbose=2,
        n_jobs=-1
    )
    grid.fit(X_smote,y_smote)
    
    print(f'Classifier: {clf}')
    print('Best Parameters:')
    print(grid.best_params_)
    
    # Scoring metrics
    scores = {'Classifier': clf}
    y_preds = grid.predict(X_test)
    y_pred_probas = grid.predict_proba(X_test)[:, 1]
    y_train_pred_probas = grid.predict_proba(X_smote)[:, 1]
    scores['Train ROC-AUC Score'] = metrics.roc_auc_score(y_smote, y_train_pred_probas)
    scores['Test ROC-AUC Score'] = metrics.roc_auc_score(y_test, y_pred_probas)
    scores['F1'] = metrics.f1_score(y_test, y_preds)
    scores['Precision'] = metrics.precision_score(y_test, y_preds)
    scores['Recall'] = metrics.recall_score(y_test, y_preds)
    scores['Accuracy'] = metrics.accuracy_score(y_test, y_preds)
    
    # Storing results
    result.append(scores)
    
    return pipe
```

```python
# Run through gridsearch for every model
for (clf, clf_params) in classifiers:
    model(clf, clf_params)
```

```python
# Results sorted by Test ROC-AUC
pd.DataFrame(result).sort_values(by='Test ROC-AUC Score', ascending=False).reset_index(drop=True)
```

Based on Gridsearch, the best model is Gradient Boosting Classifier with 87% of ROC_AUC with the best parameter of 'learning_rate': 0.15, 'max_depth': 3. <br>
And since this model performs better than baseline model and Keras with L2 regularization, we will select this as our final prediction model for this project
