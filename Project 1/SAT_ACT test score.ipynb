    The formula for standard deviation is below:

    $$\sigma = \sqrt{\frac{1}{n}\sum_{i=1}^n(x_i - \mu)^2}$$

    Where $x_i$ represents each value in the dataset, $\mu$ represents the mean of all values in the dataset and $n$ represents the number of values in the dataset.

## Data Import and Cleaning

**Read and clean SAT 2019 dataset** <br>
1) remove % from Participation rate and convert to decimals<br>
2) Composite column contains string, remove unwanted string and convert to numeric<br>
3) Create new column to compute percentage grades as ACT full marks is 1600. Thus, total score divide by 1600 to be the grade score for individual states. This is useful to make comparison with ACT test score

**Remove percentage sign and convert column into float to show values in decimals for Participation Rate**

```python
sat19['Participation Rate'] = sat19['Participation Rate'].str.strip('%').replace(r'—', 0, regex=True).astype('float')/100
```

**SAT total total result, full marks is 1600, create new column to compute the percentage score in decimal point, this will be useful when comparing ACT scores**

```python
sat19 = sat19.rename(columns=str.lower)
sat19['year'] = '2019'
sat19['sat_grades'] = (sat19['total']/1600*100).round(decimals=2)

sat19.head()
```

**Repeat above steps for SAT 2018 scores** <br>
**Note** - Did not create to repeat the clean up is because there are slight difference for each columns in individual SAT datasets.

```python
sat18 = pd.read_csv('data/sat_2018.csv')
sat18.head()
```

```python
print(sat18.isnull().sum())
sat18.dtypes
```

```python

sat18 = sat18.rename(columns={'Participation':'participation rate','Evidence-Based Reading and Writing':'ebrw'})
sat18['participation rate'] = sat18['participation rate'].str.strip('%').replace(r'—', 0, regex=True).astype('float')/100
sat18 = sat18.rename(columns=str.lower)
sat18['year'] = '2018'
sat18['sat_grades'] = (sat18['total']/1600*100).round(decimals=2)
sat18.head()
```

**Repeat above steps for SAT 2017 scores**

```python
sat17 = pd.read_csv('data/sat_2017.csv')
sat17 = sat17.rename(columns={'Participation':'participation rate','Evidence-Based Reading and Writing':'ebrw'})
sat17['participation rate'] = sat17['participation rate'].str.strip('%').replace(r'—', 0, regex=True).astype('float')/100
sat17 = sat17.rename(columns=str.lower)
sat17['year'] = '2017'
sat17['sat_grades'] = (sat17['total']/1600*100).round(decimals=2)
sat17.head()
```

**Append all SAT test for 2017 to 2019 into dataframe**

**Remove rows that has zero participation rate from our analysis**

**Read and clean ACT 2017 dataset**<br>
1) remove % from Participation rate and convert to decimals<br>
2) Composite column contains string, remove unwanted string and conver to numeric<br>
3) Create new column to compute percentage grades as ACT full marks is 36. Thus, composite score divide by 36 to be the grade score for individual states. This is useful to make comparison with SAT test score

```python
act17 = pd.read_csv('data/act_2017.csv')
act17['Participation'] = act17['Participation'].str.strip('%').replace(r'—', 0, regex=True).astype('float')/100
act17 = act17.rename(columns={'Participation':'act_participation'})
act17 = act17.rename(columns=str.lower)
act17['year'] = '2017'
act17['composite'] = act17['composite'].replace(r'x', 0, regex=True)
act17['composite'] = act17['composite'].astype(float)
act17['act_grades'] = (act17['composite']/36*100).round(decimals=2)
act17.head()
```

**Read and clean ACT 2018 dataset**<br>
Repeat the steps above

```python
act18 = pd.read_csv('data/act_2018.csv')
act18['Participation'] = act18['Participation'].str.strip('%').replace(r'—', 0, regex=True).astype('float')/100
act18 = act18.rename(columns={'Participation':'act_participation'})
act18 = act18.rename(columns=str.lower)
act18['year'] = '2018'
act18['act_grades'] = (act18['composite']/36*100).round(decimals=2)
act18.head()
```

**Read and clean ACT 2019 dataset**<br>
Repeat the steps above

```python
act19 = pd.read_csv('data/act_2019.csv')
act19['Participation'] = act19['Participation'].str.strip('%').replace(r'—', 0, regex=True).astype('float')/100
act19 = act19.rename(columns={'Participation':'act_participation'})
act19 = act19.rename(columns=str.lower)
act19['year'] = '2019'
act19['act_grades'] = (act19['composite']/36*100).round(decimals=2)
act19.head()
```

**Append all ACT dataset into one Dataframe from 2017 to 2019**<br>
**Assumption** - states that have zero grades but has participation rate deemed to be irrelevant. Thus, the decision to drop the row.

```python
act = act17.append([act18,act19])

act['state'] = act['state'].str.lower()
act['state'] = act['state'].str.capitalize()
act = act[~(act['act_grades']==0)]
```

**Concatenate both ACT and SAT Dataset from 2017 to 2019 and drop unnecessary columns for comparison**

```python
df = act.merge(right=sat,on = ['state','year'], how='outer')
df.drop(columns=['ebrw','math_x','math_y','english','total','reading','science','composite'],inplace=True)

```

**Assumption** - Student who took ACT test may not necessarily take SAT test.<br>
**Note** - for states that have students taking both tests, will sum up both participation rate and grades and divide by 2 to get the average scores and participation. Otherwise, will take either one of the score and participation rate.

```python
df.replace(np.nan,0)
df.loc[(df['act_participation'] == 0) | (df['sat_participation'] == 0),'participation'] = df['act_participation'] + df['sat_participation']
df.loc[(df['act_participation'] > 0) & (df['sat_participation'] > 0),'participation'] = round((df['act_participation'] + df['sat_participation'])/2,2) 

df.loc[(df['act_grades'] == 0) | (df['sat_grades'] == 0),'grades'] = df['act_grades'].replace(np.nan,0) + df['sat_grades']
df.loc[(df['act_grades'] > 0) & (df['sat_grades'] > 0),'grades'] = round((df['act_grades'].replace(np.nan,0) + df['sat_grades'])/2,2) 
df
```

## Exploratory Data Analysis

Create scatter graph to illustrate the correlation and the trend between Participation rate and Grades from 2017 to 2019

```python
sns.lmplot(x='participation', y='grades', data=df, hue='year', ci=False)
plt.xlabel('Participate Rate')
plt.ylabel('Grades');
plt.title('Avg ACT/SAT grade vs Partipication Rate by US State' );
```

**Assumption** - the higher the grade the lower the participation rate is.<br>
Thus, to better illustrate the correlation, rearrange the grades by desending orders 

```python
stat = df[df['grades'] > 70]
stat.sort_values(by=['grades','year'],ascending=False)
```

```python
group = df.groupby('state').mean().round(2)
group.sort_values(by='grades',ascending=False)[0:5]
```

**Observation** -  by comparing ACT with low participate rate yields higher ACT grades than higher SAT Participate rate with lower SAT_grades and this applies to all other states as well<br>
In other words, when states forces students to take the test, it will inevitably pull down the grades. the reason is simple, it is because when states forces involuntary students, they tend not to perform well with their test. Therefore, it pulls down the average grades in the states. <br>
In contrast, when a motivated student take the test without enforcement, they tend to perform better which push the state's average grades to be higher.

```python
group.sort_values(by='grades',ascending=True)[0:5]
```
